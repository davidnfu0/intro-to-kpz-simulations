# Semicircle Law Simulations Playground
In this notebook, we will simulate the semicircle law via the eigenvalues of random matrices, GOE and GUE.

```python
from semicirclelaw import create_animation_semicircle_law
from IPython.display import HTML
```

### Setting up

```python
SEED = 97
NS = [10, 20, 50, 100, 200, 500, 1000, 2000, 4000, 6000, 10000]
INTERVAL = 600
```

### Animations

```python
anim = create_animation_semicircle_law(
    Ns=NS, matrix_type="GOE", seed=SEED, interval=INTERVAL
)
HTML(anim.to_jshtml())
```

```python
anim = create_animation_semicircle_law(
    Ns=NS,
    matrix_type="GUE",
    seed=SEED,
    interval=INTERVAL,
)
HTML(anim.to_jshtml())
```
